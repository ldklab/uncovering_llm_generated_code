```python
import pickle
apps_scores = pickle.load(open("roc_aucs_apps_codellama_rewrite_m32.pkl","rb"))
mbpp_scores = pickle.load(open("roc_aucs_mbpp_codellama_rewrite_m32.pkl","rb"))
```

```python
import matplotlib.pyplot as plt
xticks = [2,4,8,12,16,20,32]
plt.figure(figsize=(8,4))
plt.rc('font',family='Noto Mono',size=16,weight='bold')
plt.plot(list(range(1,33)),[item for item in apps_scores], label='APPS')
plt.plot(list(range(1,33)),[item for item in mbpp_scores], label='MBPP')
plt.xlabel('Number of Rewriting: m',size=16,weight='bold')
plt.xticks(xticks)
ax = plt.gca()
ax.grid(False)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.legend()
plt.savefig("visuals/impact_of_m.pdf",dpi=300,format='pdf')
```
