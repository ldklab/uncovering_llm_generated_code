```python
%load_ext autoreload
%autoreload 2
import os
import numpy as np
os.chdir("../")
```

```python
import pandas as pd
hum_test_data = pd.read_pickle("dataset/apps_hum_test.pkl")
gen_test_data = pd.read_pickle("dataset/apps_gen_test.pkl")
all_data = pd.concat([hum_test_data,gen_test_data])
```

```python
all_data = all_data[all_data['extracted_full_func'].apply(len) > 0]
```

```python
import numpy as np
def get_avglogprob(x):
    if len(x) == 0:
        return 0
    return np.mean(np.log(np.array(x) + 1e-7))

score = all_data['prompt+extracted_func_word_prob_vicuna-13b'].apply(get_avglogprob)
label = all_data['label']
```

```python
from sklearn.metrics import average_precision_score,roc_auc_score
roc_auc = roc_auc_score(label,score)
ap = average_precision_score(label,score)
roc_auc,ap
```
