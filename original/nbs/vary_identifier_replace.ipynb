```python
import matplotlib.pyplot as plt

names = ['logProb','Rank','LogRank','Entropy','DetectGPT','Ours']

apps_orig = [0.610,0.513,0.590,0.570,0.618,0.785]
apps_ident01 = [0.565,0.4725,0.5375,0.5475,0.584,0.741]
apps_ident02 = [0.545,0.4625,0.52,0.5375,0.558,0.6457]
apps_ident05 = [0.515,0.4725,0.5075,0.515,0.512,0.520]

fill_colors = ['green', 'blue', 'purple', 'orange', 'red', 'yellow']
edge_colors = ['black'] * 6
hatches_01 = '..'
hatches_02 = 'xx'
hatches_05 = '//'
x_center = list(range(6))
bar_width = 0.3
bar_height = 2


# plt.rc('font', family='Times New Roman', size=10)
plt.figure(figsize=(8,3),dpi=100)
plt.subplot(1,1,1)
plt.bar([item-bar_width for item in x_center],apps_ident01, color=fill_colors, width=bar_width, hatch=hatches_01, edgecolor=edge_colors)

plt.bar([item for item in x_center],apps_ident02, color=fill_colors, width=bar_width, hatch=hatches_02, edgecolor=edge_colors)

plt.bar([item+bar_width for item in x_center],apps_ident05, color=fill_colors, width=bar_width, hatch=hatches_05, edgecolor=edge_colors)
plt.xticks(x_center,names)

ax=plt.gca()
ax.set_title("Detection On APPS",font='Noto Mono',size=16,fontweight='heavy')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.set_ylim([0.4, 0.9])
for tick in ax.get_xticklabels():
    tick.set_rotation(45)
    tick.set_horizontalalignment('right')
    tick.set_fontname('Noto Mono')
    tick.set_fontsize(12)
# plt.grid(True)

# plt.savefig("visuals/vary_generation_prompt.pdf",format="pdf",bbox_inches="tight")
```
